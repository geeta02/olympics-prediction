```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
file_path = "/Users/geetharajamanickam/Downloads/DS Group Project/phdi_yearwise_avg.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
df.info()
```

```python
print('Summary Statistics: ')
df.describe()
```

```python
print("Skewness: ")
df.iloc[:, 2:].skew()
```

```python
print("Kurtosis: ")
df.iloc[:, 2:].kurt()
```

```python
corr_features = df.drop(columns = ['iso3', 'country'])
corr_matrix = corr_features.corr()
sns.heatmap(corr_matrix, cmap = 'Blues', annot = True)
plt.show()
```

```python
num_cols = ['1996-2000', '2000-2004', '2004-2008', '2008-2012']
sns.set_style("whitegrid")
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

for i, col in enumerate(num_cols):
    sns.histplot(df[col].dropna(), bins=30, kde=True, ax=axes[i], color='skyblue')
    axes[i].set_title(f"Histogram of {col}")
    axes[i].set_xlabel("Value")
    axes[i].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
plt.figure()
sns.boxplot(data=df[num_cols], palette="Set2")
plt.title("Boxplot of MMR values across years")
plt.xlabel("Year")
plt.ylabel("Value")
plt.show()
```
