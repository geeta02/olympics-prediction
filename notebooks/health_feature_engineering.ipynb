# BMI

```python
# importing the dataset
BMI_raw_data = pd.read_csv(r'BMI_dataset.csv')
BMI_raw_data
```

```python
BMI_raw_data = BMI_raw_data[["Year", "Sex", "ISO", 
                              "Prevalence of BMI<18.5 kg/m² (underweight)", 
                              "Prevalence of BMI>=30 kg/m² (obesity)",
                              "Prevalence of BMI >=40 kg/m² (morbid obesity)"]]

BMI_raw_data
```

```python
BMI_data = (
    BMI_raw_data
    .groupby(["Year", "ISO"], as_index=False)
    .agg({
        "Prevalence of BMI<18.5 kg/m² (underweight)": "mean",
        "Prevalence of BMI>=30 kg/m² (obesity)": "mean",
        "Prevalence of BMI >=40 kg/m² (morbid obesity)": "mean"
    })
)

# renaming columns
BMI_data.rename(columns={"Prevalence of BMI<18.5 kg/m² (underweight)" : "BMI_underweight",
                        "Prevalence of BMI>=30 kg/m² (obesity)" : "BMI_obesity",
                        "Prevalence of BMI >=40 kg/m² (morbid obesity)" : "BMI_morbid_obesity"}, inplace=True)

BMI_data
```

```python
BMI_data = BMI_data
```

# Height

```python
# importing the dataset
height_raw_data = pd.read_csv(r'Height_dataset.csv')
height_raw_data
```

```python
height_raw_data = height_raw_data[["Country", "Sex", "Year", "Age group", "Mean height"]]
height_raw_data
```

```python
height_data = height_raw_data[height_raw_data["Age group"].between(17, 19, inclusive="both")]
height_data
```

```python
# Group by (country, year) and average the 'height' column
height_data = (
    height_data
    .groupby(["Year", "Country"])["Mean height"]
    .mean()                          # This aggregates across all sexes
    .reset_index()
)

# renaming columns
height_data.rename(columns={"Mean height" : "Mean_Height"}, inplace=True)

height_data
```

```python
import country_converter as coco # handles messy country codes

height_data['ISO'] = coco.convert(names=height_data['Country'], to='ISO3')
height_data = height_data[["Year","ISO","Mean_Height"]]
height_data
```

# Diabetes

```python
# importing the dataset
diabetes_raw_data = pd.read_csv(r'Diabetes_dataset.csv')
diabetes_raw_data.head()
```

```python
# selecting the wanted features
diabetes_data = diabetes_raw_data[["ISO", "Sex", "Year", "Prevalence of diabetes (18+ years)", 
                                       "Proportion of people with diabetes who were treated (30+ years)"]]


# renaming the fearures
diabetes_data = diabetes_data.rename(columns={"Prevalence of diabetes (18+ years)" : "Diabetes_in_18+",
                             "Proportion of people with diabetes who were treated (30+ years)" : "Diabetes_treated_in_30+"})

diabetes_data
```

```python
# grouping the Sex column into one

# diabetes_data = (diabetes_data.groupby(["Year", "ISO"]).mean().reset_index())

# Suppose your DataFrame is named 'diabetes_data'
# and it contains a "Sex" column along with numeric columns.
# We'll group by 'Year' and 'ISO' (country code), ignoring the 'Sex' column.
# Group by Year and country # Take the mean of numeric columns
diabetes_data = (diabetes_data.groupby(["Year", "ISO"], as_index=False).mean(numeric_only=True))
diabetes_data
```

# Merge

```python
import pandas as pd
from functools import reduce

datasets = [BMI_data, height_data, diabetes_data, cholesterol_data, blood_pressure_data, hypertension_data]

def merge_on_year_iso(left, right):
    return pd.merge(left, right, on=["Year","ISO"], how="inner")

df_final = reduce(merge_on_year_iso, datasets)

df_final
```

```python
output_filename = "anthropometric_data.csv"

df_final.to_csv(output_filename, index=False)

print(f"Exported final dataset to {output_filename}")
```
